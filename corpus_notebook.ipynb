## Import corpus from all files in the folder

```python
# import corpus from all files in the folder
i = 0
corpus = []
fnames = []
# load corpus
for f in os.listdir():
    doc = [line.rstrip('\n') for line in open(f)]
    print(f)
    corpus.append(doc)
    fnames.append(os.path.basename(f))
    i += 1
Nfiles = i-1
# declare a function intersection
def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return lst3
# check a random pair of data
intersection(corpus[23],corpus[5])
```
